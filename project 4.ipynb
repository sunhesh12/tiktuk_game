```python
board={
    "P1":" ","P2":" ","P3":" ",
    "P4":" ","P5":" ","P6":" ",
    "P7":" ","P8":" ","P9":" "
}
player=1
Count_moves=0
win=False

def check_win():
    global win

    #Horozontal
    if(all(board[k]=="x" for k in["P1","P2","P3"])) or (all(board[k]=="x" for k in["P4","P5","P6"])) or (all(board[k]=="x" for k in["P7","P8","P9"])):
        print("Player 01 win the Game(-)")
        win=True
    if(all(board[k]=="o" for k in["P1","P2","P3"])) or (all(board[k]=="o" for k in["P4","P5","P6"])) or (all(board[k]=="o" for k in["P7","P8","P9"])):
        print("Player 02 win the Game(-)")
        win=True    

    #Vertical
    if(all(board[k]=="x" for k in["P1","P4","P7"])) or (all(board[k]=="x" for k in["P2","P5","P8"])) or (all(board[k]=="x" for k in["P3","P6","P9"])):
        print("Player 01 win the Game(-)")
        win=True
    if(all(board[k]=="o" for k in["P1","P4","P7"])) or (all(board[k]=="o" for k in["P2","P5","P8"])) or (all(board[k]=="o" for k in["P3","P6","P9"])):
        print("Player 02 win the Game(-)")
        win=True

    #Diagonal
    if(all(board[k]=="x" for k in["P1","P5","P9"])):
        print("Player 01 win the Game(\)")
        win=True
    if(all(board[k]=="o" for k in["P1","P5","P9"])):
        print("Player 02 win the Game(\)")
        win=True 
    if(all(board[k]=="x" for k in["P3","P5","P7"])):
        print("Player 01 win the Game(/)")
        win=True
    if(all(board[k]=="o" for k in["P3","P5","P7"])):
        print("Player 02 win the Game(/)")
        win=True   


print("Let's start the game")

while True:
    print(" __ __ __")
    print("|"+board["P1"]+" |"+board["P2"]+" |"+board["P3"]+" |")
    print("|--|--|--|")
    print("|"+board["P4"]+" |"+board["P5"]+" |"+board["P6"]+" |")
    print("|--|--|--|")
    print("|"+board["P7"]+" |"+board["P8"]+" |"+board["P9"]+" |")
    

    check_win()
    if Count_moves==9 or win:
        break
        
    while True:
        if player==1:
            P1=input("Player 01, Select the position: ")
        
            if(P1.upper() in board) and (board[P1.upper()]==" "):
                board[P1.upper()]="x"
                player=2
                
                break
            else:
                print("position is not available")
                continue   
        else:
            P2=input("Player 02, Select the position: ")
        
            if(P2.upper() in board) and (board[P2.upper()]==" "):
                board[P2.upper()]="o"
                player=1
                
                break
            else:
                print("position is not available")
                continue 
        
    Count_moves+=1
    
print("")
print("")
print("Game end...")  
```
